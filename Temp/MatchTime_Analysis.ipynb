```python
#ATP
import numpy as np
import pandas as pd
yrstart=2017
yrend=2021
df=pd.DataFrame()
for year in range(yrstart, yrend+1):
    dfcha=pd.read_csv((r"https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_qual_chall_{0}.csv").format(str(year)),index_col=0)
    dfatp=pd.read_csv((r"https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_{0}.csv").format(str(year)),index_col=0)
    df=pd.concat([df,dfcha,dfatp],axis=0)

df["surface2"]=df["surface"].map({'Hard':'NonClay','Grass':'NonClay','Carpet':'NonClay','Clay':'Clay'})
court=['NonClay','Clay']
level=['A','C']
df["tourney_level2"]=df["tourney_level"]
#A=ATP250/500,C=CH,D=DC,F=ATP FINALS,G=GS,M=M1000
df["tourney_level2"]=df["tourney_level2"].replace(['A','D','M','G','F'],'Atp')
df["tourney_level2"]=df["tourney_level2"].replace(['C'],'Ch')
```

```python
mask = df["round"].str.startswith('Q')
df.loc[mask, 'tourney_level2']=df[mask].replace(['Atp','Ch'],['AtpQ','ChQ'])
df.head(10)
```

```python
df["seconds_per_pt"]=df["minutes"]*60/(df["w_svpt"]+df["l_svpt"]).round(decimals = 0)
#add a Field that indicated if the row has the player inverted (for scores, stats...)
df["isRevertedPlayers"]=False
```

```python
#create a copy of the dataframe but with players inverted
df_reverted_players=df.copy()
df_reverted_players["isRevertedPlayers"]=True
df_reverted_players[["w_ace","l_ace"]]=df_reverted_players[["l_ace","w_ace"]]
df_reverted_players[["winner_name","loser_name"]]=df_reverted_players[["loser_name","winner_name"]]
df_reverted_players.head()
```

```python
df_with_rev_players=pd.concat([df, df_reverted_players], axis=0)
df_with_rev_players["w_ace_rate"]=(df_with_rev_players["w_ace"]/df_with_rev_players["w_svpt"]).round(decimals = 3)
df_with_rev_players["l_ace_rate"]=(df_with_rev_players["l_ace"]/df_with_rev_players["l_svpt"]).round(decimals = 3)

df_with_rev_players.head()
```

```python
field_to_eval="w_ace_rate"#"seconds_per_pt"
df_by_players=df_with_rev_players.groupby(by=["winner_name","surface","tourney_level2"]).agg(["median","mean","count"])
df_by_players=df_by_players[df_by_players[(field_to_eval,"count")]>20]
my_query=("surface=='{0}'").format(court)
#df_by_players=df_by_players.query(my_query)
df_by_players=df_by_players.sort_values([(field_to_eval,"median")])
df_by_players
```

```python
###
#LOOK for ace rate between ATP and CH level
#TRY to rate how easier it is to hit an ace at CH level comparing to ATP level
#for the players who play at both levels, check the diff of ace_rate
###
df_ATPvsCH=df_by_players.loc[:,field_to_eval][["mean","median","count"]]
my_query=("(tourney_level2=='Atp' | tourney_level2=='Ch') & surface=='Hard' ")
df_ATPvsCH=df_ATPvsCH.query(my_query, engine='python')#.loc[:,field_to_eval][["median","mean","count"]]
df_ATPvsCH=df_ATPvsCH.reset_index()
#Filter to get only the rows when a player has a ROW for ATP AND a Row for Ch
df_name_is_twice=df_ATPvsCH['winner_name'].map(df_ATPvsCH['winner_name'].value_counts()) > 1
df_ATPvsCH=df_ATPvsCH[df_name_is_twice].sort_values(by=["mean"])
print(df_ATPvsCH.tail(20))
print(df_ATPvsCH[df_ATPvsCH.tourney_level2=="Atp"]["mean"].count())
print(df_ATPvsCH[df_ATPvsCH.tourney_level2=="Ch"]["mean"].count())
print(df_ATPvsCH[df_ATPvsCH.tourney_level2=="Ch"]["mean"].sum()/df_ATPvsCH[df_ATPvsCH.tourney_level2=="Atp"]["mean"].sum())
#Now we must turn the 2 lines for a player (1 for ATp, 1 for CH into 1 line with multiple columns:
df_ATPvsCH_byplay=pd.pivot_table(df_ATPvsCH,index="winner_name",columns="tourney_level2", aggfunc=np.sum)
df_ATPvsCH_byplay["median_diff"]=df_ATPvsCH_byplay["median"]["Ch"]/df_ATPvsCH_byplay["median"]["Atp"]
df_ATPvsCH_byplay["mean_diff"]=df_ATPvsCH_byplay["mean"]["Ch"]/df_ATPvsCH_byplay["mean"]["Atp"]
df_ATPvsCH_byplay=df_ATPvsCH_byplay.sort_values(by=[('median', 'Atp')], ascending=False)
print(df_ATPvsCH_byplay.head(50))
#df_ATPvsCH_final
print
```
