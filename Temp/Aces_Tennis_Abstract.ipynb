```python
import numpy as np
import pandas as pd
df=pd.read_csv(r"https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv",index_col=0)
```

```python
df=pd.concat([df, df], axis=0)
df.head()
```

```python
df.info()
```

```python
#get a list of all distinct players
list_players=list(df.loser_name)+list(df.winner_name)
list_players=list(set(list_players))
list_players.sort()
```

```python
#GROUP BY player and surface (for the winners of the matches)
df_aces_by_ply1_surf=df.pivot_table(values=['l_ace','w_1stWon','w_2ndWon','w_ace','w_svpt'], index=['winner_name','surface'],\
                    aggfunc=np.sum)
df_aces_by_ply1_surf.columns=['ace_vs','1stWon','2ndWon','ace','svpt']
df_aces_by_ply1_surf=df_aces_by_ply1_surf.rename_axis(['name', 'surface'])
df_aces_by_ply1_surf
```

```python
#GROUP BY player and surface (for the losers of the matches)
df_aces_by_ply2_surf=df.pivot_table(values=['l_1stWon','l_2ndWon','l_ace','l_svpt','w_ace','w_svpt'], index=['loser_name','surface'],\
                    aggfunc=np.sum)
df_aces_by_ply2_surf.columns=['1stWon','2ndWon','ace','svpt','ace_vs']
df_aces_by_ply2_surf.rename(columns={'l_1stWon' : '1stWon', 'l_2ndWon' : '2ndWon','l_ace':'ace','l_svpt':'svpt','w_ace':"ace_vs",'w_svpt':'svpt_vs'}, inplace=True)
df_aces_by_ply2_surf=df_aces_by_ply2_surf.rename_axis(['name', 'surface'])
df_aces_by_ply2_surf
```

```python
#add up both data frames and count 0 if one of them has no value
df_aces_by_ply_surf = df_aces_by_ply1_surf.add(df_aces_by_ply2_surf, fill_value=0)
```

```python
df_aces_by_ply_surf
```

```python
df_aces_by_ply_surf["svpt_W"]=df_aces_by_ply_surf["1stWon"]+df_aces_by_ply_surf["2ndWon"]
df_aces_by_ply_surf["svpt_W"]=df_aces_by_ply_surf["1stWon"]+df_aces_by_ply_surf["2ndWon"]
df_aces_by_ply_surf["aceW_pt"]=df_aces_by_ply_surf["ace"]/df_aces_by_ply_surf["svpt_W"]
df_aces_by_ply_surf["aceL_pt"]=df_aces_by_ply_surf["ace_vs"]/df_aces_by_ply_surf["svpt_W"]
df_aces_by_ply_surf.head()
```
